```python
for j in range(10):
    hists = data.patch_grandhist([i+1 for i in range(100)],j,3)
    fig,ax = plt.subplots(3)
    color = ['red','green','blue']
    [ax[i].bar(hists[i][1][:-1],hists[i][0],color = color[i]) for i in range(3)]
    [ax[i].set_title('Image Histogram for '+color[i]+' color channel: '+data.part_dict[j]) for i in range(3)]
    ax[2].set_xlabel('Intensity')
    ax[1].set_ylabel('Relative Frequency')
    plt.tight_layout()
    plt.savefig('../figures/'+'Image Histogram for '+data.part_dict[j]+'.png')
    plt.show()
```

```python
for j in range(10):
    hists = data.patch_grandhist([i+1 for i in range(100)],j,3)
    fig,ax = plt.subplots(3)
    color = ['red','green','blue']
    [ax[i].bar(hists[i][1][:-1],hists[0][0]/np.sum(hists[i][0]).astype(float),color = color[i]) for i in range(3)]
    [ax[i].set_title('Image Histogram for '+color[i]+' color channel: '+data.part_dict[j]) for i in range(3)]
    ax[2].set_xlabel('Intensity
                     ')
    ax[1].set_ylabel('Relative Frequency')
    plt.tight_layout()
    plt.show()
```

```python
all_hists = data.hists_wholemouse([i+1 for i in range(100)],0,bins = np.linspace(0,80,81))
for j in range(5):
    for i in range(j):
        plt.bar(all_hists[(j,i)][1][:-1],all_hists[(j,i)][0])
        plt.title('Distance between '+data.part_dict[j]+' and '+data.part_dict[i])
        plt.xlabel('Norm (Pixels)')
        plt.ylabel('Count')
        plt.show()
all_hists = data.hists_wholemouse([i+1 for i in range(100)],1,bins = np.linspace(0,80,81))
for j in range(5):
    for i in range(j):
        plt.bar(all_hists[(j+5,i+5)][1][:-1],all_hists[(j+5,i+5)][0])
        plt.title('Distance between '+data.part_dict[j+5]+' and '+data.part_dict[i+5])
        plt.xlabel('Norm (Pixels)')
        plt.ylabel('Count')
        plt.show()
```
